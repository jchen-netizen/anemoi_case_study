```python
import numpy as np 
import xarray as xr
# import netCDF4 as nc
# from datetime import datetime, timedelta
```

```python
ds = xr.open_dataset('../wrfout_d02_2023-05-11_00_00_00')
```

```python
# Variable
var = "T2" # 2-meter temperature

# Pink Mountain coordinates
lat_pink = 56.94
lon_pink = -122.7
alt_pink = 960.10

# Find the nearest grid point to the Pink Mountain location
dist = np.sqrt((ds["XLAT"] - lat_pink)**2 + (ds["XLONG"] - lon_pink)**2) # compute distance to every grid point
min_idx = np.unravel_index(np.argmin(dist.values), dist.shape) # index of nearest point
sn_idx = min_idx[1]
we_idx = min_idx[2]
nearest_lat = float(ds["XLAT"].values[min_idx]) # nearest grid point coordinates
nearest_lon = float(ds["XLONG"].values[min_idx])

# Extract the variable value at the nearest grid point
var_value = ds[var].isel(south_north=sn_idx, west_east=we_idx)

print("nearest index:", min_idx)
print("nearest lat/lon:", nearest_lat, nearest_lon)
print(f"{var} at Pink Mountain: {var_value.values}")
print(f"{var_value.values - 273.15} °C") # convert from K to °C
```
